```python
#Answer of 1
import numpy
import matplotlib.pyplot as plt

N=700

x= numpy.random.uniform(low=-1, high=1, size=[N,1])
y= numpy.random.uniform(low=-1,  high=1, size=[N,1])

inside_bool = x**2+y**2<1

approx_pi = 4*numpy.sum(inside_bool)/N
print('Pi: {}, approximation: {}'.format(numpy.pi, approx_pi))

x_in = x[inside_bool]
y_in = y[inside_bool]

plt.figure(figsize=[2,2])
plt.scatter(x, y , s=1)
plt.scatter(x_in, y_in, color='r', s=1)
plt.show()
```

```python
#Answer of 2
import random
# defining function with name and number of flips argument
def heads_tails(number_of_flips):
    # counting variables to hold tally count (will use with addition compound assigment operator)
    tails_count = 0
    heads_count = 0
    distance=0
    # using loop to flip coin specified number of times using number passed when function called
    for i in range(number_of_flips):
        # each time through loop or for each coin flip generate random integer of 1 or 2
        rand = random.randint(1,2)
        # if random integer 1 this will represent a tail and adding 1 to tails count
        if rand == 1:
            tails_count += 1
            # printing tails tally 
            print(tails_count, 'tails')
        # if random integer is 2 this will represent a head and adding 1 to heads count
        else:
            heads_count += 1
            # printing heads tally 
            print(heads_count, 'heads')
            distance=heads_count-tails_count
    # printing final tally count for all coin flips
    print('tails', tails_count)
    print('heads', heads_count)
    print(distance)
    # call function to simulate n coin flips    
heads_tails(25)



```
